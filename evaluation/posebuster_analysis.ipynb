```python
import glob
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
"""
Conformation generation
"""
conf_path = 'conf_buster'
csv_ls = Path(conf_path).glob('*.csv')
result_df = pd.DataFrame()
for csv in csv_ls:
    df = pd.read_csv(str(csv))
    #if want to check whether the file was loaded properly, replace 2 with 1
    metrics_df =pd.DataFrame(df.iloc[:,2:].mean(axis=0)).T 
    metrics_df.insert(loc=0, column='Filename', value=(csv.name))
    result_df = pd.concat([result_df, metrics_df])
result_df
```

```python
"""
Pocket generation
"""
check_metrics = [
    'sanitization',
    'bond_lengths',
    'bond_angles',
    'internal_steric_clash',
    'internal_energy',
    'protein-ligand_maximum_distance',
    'minimum_distance_to_protein',
    'volume_overlap_with_protein'
    ]

conf_path = 'dock_conf/epo20_100steps'
csv_ls = Path(conf_path).glob('*.csv')
result_df = pd.DataFrame()
for csv in csv_ls:
    df = pd.read_csv(str(csv))[check_metrics].replace(True, 1).replace(False, 0)
    metrics_df =pd.DataFrame(df.mean(axis=0)).T 
    result_df = pd.concat([result_df, metrics_df])
pd.DataFrame(result_df.mean(axis=0)).T
```
